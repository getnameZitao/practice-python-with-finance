I will try out the MC (monte carlo) simulation in this note page, goal is to read along and simulate the equations mentioned in the book here using python

```python
#this code is from chatgpt
import numpy as np
import matplotlib.pyplot as plt

# Parameters
S0 = 100        # Initial stock price
r = 0.05        # Risk-free rate
sigma = 0.2     # Volatility
T = 1.0         # Time horizon (1 year)
N = 252         # Number of time steps (trading days)
dt = T / N      # Time step size

# Initialize stock price path
prices = [S0]
for i in range(N):
    z = np.random.normal()  # Random draw from standard normal distribution
    St = prices[-1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * z)
    prices.append(St)

# Plot the simulated price path
plt.figure(figsize=(10, 5))
plt.plot(prices)
plt.title("Simulated Stock Price Path (GBM)")
plt.xlabel("Time Step (Days)")
plt.ylabel("Stock Price")
plt.grid(True)
plt.show()
```

```python
#import stacks
import math
import numpy as np
#define the variables in BSM
S0 = 100
K = 105
T = 1.0
r = 0.05
sigma = 0.2
I=1000
#generate the random number for MC sim.
np.random.seed(1000)
#define the randomness as a standarn distribution
z = np.random.standard_normal(I)
# equation to calculated strike price at time T
ST = S0*np.exp((r - sigma ** 2 / 2) * T + sigma * math.sqrt(T) * z)
#the option payoff at time T
hT = np.maximum(ST - K, 0)
#the MC estimator
C0 = math.exp(-r * T) * np.mean(hT)
print('Value of the EU call option: %5.3f'% C0)
```
